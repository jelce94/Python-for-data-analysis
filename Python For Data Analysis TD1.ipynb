### Find and answer 3 questions to ask using pandas, including aggregates.

```python
# Top 5 des clubs de premier league ayant le plus marqué durant la saison 2018/2019
top_clubs_goals = premier_league.groupby("Current Club").goals_overall.sum().sort_values(ascending=False)[:5]
top_clubs_goals
```

```python
# Nombre de milieux de terrains par club classé par ordre croissant
nb_midfielder = premier_league[premier_league["position"] == 'Midfielder'].groupby("Current Club").league.count().sort_values()
nb_midfielder
```

```python
# Nationalité des joueurs la plus représentée en premier league
top_nationality = premier_league.groupby('nationality').league.count().sort_values(ascending=False)
top_nationality
```

### Find and display 3 plots using matplotlib, including pivot tables.

```python
# Graphique montrant le nombre de buts à l'extérieur et à domicile pour chaque club 
new_df = premier_league[['Current Club', 'goals_overall', 'position']]
new_df = pd.pivot_table(new_df, values='goals_overall', index='Current Club', columns='position')
new_df
```

```python
plt.rcParams["figure.figsize"] = [15,6]
plt.rcParams["figure.dpi"] = 300
plt.xticks(rotation = 'vertical')
plt.xlabel('Clubs')
plt.ylabel('Number of goals')
plt.title("Number of goals per position for each club")
plt.plot(new_df[['Goalkeeper','Defender','Midfielder','Forward']])
```

```python
# Influence de l'age sur le nombre de min joués
total_minutes = premier_league.pivot_table('min_per_match', index='age', columns='position',aggfunc=sum)
total_minutes
total_minutes.drop(0, inplace=True)
```

```python
total_minutes.plot(title='Nombre de minutes jouées en fonction de l\'age')
```

```python
# nombre de buts concédés du top 50 des defenseurs par nombre de carton rouge pris
Penalty = premier_league.pivot_table('conceded_overall', index='rank_in_league_top_defenders', columns='red_cards_overall')
Penalty[:50]
```

```python
Penalty.plot(title="nombre de buts concédés du top 50 des defenseurs par nombre de carton rouge pris")
```
